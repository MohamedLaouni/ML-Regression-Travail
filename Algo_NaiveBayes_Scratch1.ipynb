```python
# ================================
# DONNÉES
# ================================
data = [
    ("Sunny",    "Hot",  "High",   "Weak",   "No"),
    ("Sunny",    "Hot",  "High",   "Strong", "No"),
    ("Overcast", "Hot",  "High",   "Weak",   "Yes"),
    ("Rain",     "Mild", "High",   "Weak",   "Yes"),
    ("Rain",     "Cool", "Normal", "Weak",   "Yes"),
    ("Rain",     "Cool", "Normal", "Strong", "No"),
    ("Overcast", "Cool", "Normal", "Strong", "Yes"),
    ("Sunny",    "Mild", "High",   "Weak",   "No"),
    ("Sunny",    "Cool", "Normal", "Weak",   "Yes"),
    ("Rain",     "Mild", "Normal", "Weak",   "Yes"),
    ("Sunny",    "Mild", "Normal", "Strong", "Yes"),
    ("Overcast", "Mild", "High",   "Strong", "Yes"),
    ("Overcast", "Hot",  "Normal", "Weak",   "Yes"),
    ("Rain",     "Mild", "High",   "Strong", "No"),
]

# ================================
# ÉTAPE 1 : compter Yes et No
# ================================
yes_total = 0  # compteur de nb totale de yes
no_total  = 0  ## compteur de nb totale de non

for row in data:
    if row[4] == "Yes":   # condition si la ligne[colonne] =yes anccrementer yes_total
        yes_total += 1
    else:
        no_total += 1     # si non anccrementer non_totale

print(f"Yes: {yes_total}  |  No: {no_total}") # afficher les valeures de totale de yes et non

# ================================
# ÉTAPE 2 : probabilités P(Yes) P(No)
# ================================
p_yes = yes_total / len(data)   # calcule la proba de P(Yes)=9/14
p_no  = no_total  / len(data)   # calcule le proba de P(No)=5/14

print(f"P(Yes) = {p_yes:.2f}")  #afficher la ruslt P(Yes)
print(f"P(No)  = {p_no:.2f}")   # afficher la ruslt P(No)

# ================================
# ÉTAPE 3 : Fonction pour calculer P(valeur | classe)
# ================================
def prob(colonne, valeur, classe):  # foction de calcule de proba conditionnele
    total_classe = 0   # nombre de yes ou non sellon la valeur donner en parametre
    total_valeur = 0   # Le nombre d'apparition de la valeur dans la colonne donnée en paramètre.

    for row in data:
        if row[4] == classe:
            total_classe += 1  # on calcule le nombre de classe(yes ou non )
            if row[colonne] == valeur:
                total_valeur += 1 # on calcule le nombre d'apparition de la valeur dans le colonne
                                  # exp colonne 1 (temperature) et la valeur ='hot' alors totale_valeur=4
    return total_valeur / total_classe # on retourne le proba

# ================================
# ÉTAPE 4 : classifier un nouveau jour
# ================================
def classifier(outlook, temperature, humidity, wind): # la fonction de classification selon le principe de Naive Bayes

    # Score pour Yes
    score_yes = p_yes  # on affecte la valeur de proba P(yes) a score yes
    score_yes *= prob(0, outlook,     "Yes") # appele a la fonction de proba avec la valeur de outlook et  la classe yes
    score_yes *= prob(1, temperature, "Yes") # appele a la fonction de proba avec la valeur de temperature et  la classe yes
    score_yes *= prob(2, humidity,    "Yes") # appele a la fonction de proba avec la valeur de humidity et  la classe yes
    score_yes *= prob(3, wind,        "Yes") # appele a la fonction de proba avec la valeur de wind et  la classe yes

    # Score pour No
    score_no = p_no
    score_no *= prob(0, outlook,     "No") #on repete la meme chose juste le changemant de classe avec Non
    score_no *= prob(1, temperature, "No")
    score_no *= prob(2, humidity,    "No")
    score_no *= prob(3, wind,        "No")

    print(f"\nScore Yes = {score_yes:.6f}")
    print(f"Score No  = {score_no:.6f}")

    if score_yes > score_no: # a la fin en prende la décision selon le plus la proba la plus grand
        print("→ Résultat : ✅ Yes (Jouer)")
    else:
        print("→ Résultat : ❌ No (Ne pas jouer)")

# ================================
# TEST
# ================================
classifier("Sunny", "Cool", "High", "Strong")
```
